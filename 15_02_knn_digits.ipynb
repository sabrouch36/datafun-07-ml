```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Load the digits dataset
digits = load_digits()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target, random_state=42
)

# Create the k-NN classifier model
knn = KNeighborsClassifier()

# Train the model using the training data
knn.fit(X_train, y_train)

# Evaluate the model accuracy on the testing data
accuracy = knn.score(X_test, y_test)
accuracy
```

```python
import matplotlib.pyplot as plt

# Create a grid of 4 rows x 6 columns to display 24 digit images
figure, axes = plt.subplots(nrows=4, ncols=6, figsize=(8, 6))

# Flatten the 2D axes array and loop through the first 24 images
for item, image, label in zip(axes.ravel(), digits.images, digits.target):
    item.imshow(image, cmap=plt.cm.gray_r)  # Display the image in grayscale
    item.set_xticks([])                     # Remove x-axis ticks
    item.set_yticks([])                     # Remove y-axis ticks
    item.set_title(label)                   # Show the digit label as title

plt.tight_layout()
plt.show()
```
